```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('../data/sample/data0/lsun/bedroom/0/0/0/000c3cbbd295fcd21994b49eb5ca69e7ef821714.jpg')
img
```

```python
img_arr = np.array(img)
img_arr = (img_arr/255)
img_arr.min(), img_arr.max()
```

```python
plt.imshow(img_arr)
```

```python
im = img_arr.copy()
imgs = [im]
ids = [5, 10, 20, 50, 100, 200, 500, 1000]
for i in range(1000):
    epsilon = np.random.normal(0, 1/2, img_arr.shape)
    im = im + epsilon
    if i in ids:
        imgs.append(im)
```

```python
fig, ax = plt.subplots(2,5, figsize = (12,5))
for i in range(2):
    for j in range(5):
        ax[i,j].imshow(imgs[i*2+j])
        ax[i,j].axis('off') 
        ax[i,j].set_title(f"noising step number {ids[1+i*2+j]}")
fig.tight_layout()
ax[0,0].set_title('original image')
plt.savefig('../report/media/noising_process.png', dpi=200)
```
